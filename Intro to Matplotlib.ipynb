```python
fig, ax=plt.subplots()
ax.plot(df['DATE'],df['MLY-CLDD-BASE45'])
ax.set_xlabel("MOnth")
ax.set_ylabel('Temp')
ax.set_title("the temp vs date")
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(df['DATE'],df['MLY-CLDD-BASE45'])
ax.set(xlabel='temp',ylabel='month',title='the year against temp')
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(df['DATE'],df['MLY-CLDD-BASE45'],marker='x',linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(2, 1,sharey=True)
ax[0].plot(df["DATE"], df["MLY-PRCP-NORMAL"],
color='b')
ax[0].plot(df["DATE"], df["MLY-PRCP-25PCTL"],
linestyle='--', color='b')
ax[0].plot(df["DATE"], df["MLY-PRCP-75PCTL"],
linestyle='--', color='b')
ax[1].plot(df1["DATE"], df1["MLY-PRCP-NORMAL"],
color='r')
ax[1].plot(df1["DATE"],df1["MLY-PRCP-25PCTL"],
linestyle='--', color='r')
ax[1].plot(df1["DATE"], df1["MLY-PRCP-75PCTL"],
linestyle='--', color='r')
ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
plt.show()
```
