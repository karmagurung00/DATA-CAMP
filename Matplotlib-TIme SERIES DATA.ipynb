```python
import pandas as pd
df = pd.read_csv('climate_change.csv',parse_dates=["date"],index_col="date")
df
```

```python
#HERE we are creating the same two graph in a same graph plot

fig, ax=plt.subplots()
ax.plot(df.index,df['co2'])
ax.plot(df.index, df['relative_temp'])

plt.show()
```

```python
#Here we are creating two different graph to two different plots

fig, ax=plt.subplots()
ax.plot(df.index,df['co2'],color='g')
ax.set(xlabel='Time')
ax.set_ylabel('co2 emission',color='g')
ax.tick_params('y',colors='g')

ax2=ax.twinx()
ax2.plot(df.index, df['relative_temp'])
ax2.set(ylabel='Relative Temperature increase')

plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(df.index, df['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```
